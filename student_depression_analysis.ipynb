```python
import numpy as np
import matplotlib.pyplot as plt

sizes=[[0 for _ in range(6)] for _ in range(5)]
cgpa=df['CGPA'].to_numpy()
wshw=df['Work/Study Hours'].to_numpy()
depression=df['Depression'].to_numpy()
for i in range(5,10):
    imin=max(0,i)
    imax=min(10,i+1)
    for j in range(0,12,2):
        jmin=max(0,j)
        jmax=min(12,j+2)
        if imax==10 and jmax==12:
            fil=(cgpa<=imax) & (cgpa>=imin) & (wshw<=jmax) & (wshw>=jmin)
        elif imax==10 and jmax!=12:
            fil=(cgpa<=imax) & (cgpa>=imin) & (wshw<jmax) & (wshw>=jmin)
        elif imax!=10 and jmax==12:
            fil=(cgpa<imax) & (cgpa>=imin) & (wshw<=jmax) & (wshw>=jmin)
        else:
            fil=(cgpa<imax) & (cgpa>=imin) & (wshw<jmax) & (wshw>=jmin)
        dep=depression[fil]
        sizes[i-5][int(j/2)]=sum(dep==True)/len(dep)
        
depression_scores=np.array(sizes).T
gpas=['5-6', '6-7','7-8','8-9','9-10']
hours_of_study=['0-2','2-4','4-6','6-8','8-10','10-12']

sizes=np.array(sizes).T
# Create the plot
fig, ax = plt.subplots()

# Display the depression scores as a grid of numbers
cax = ax.matshow(depression_scores, cmap="coolwarm", alpha=0.3)

# Add the numbers inside the grid
for i in range(len(hours_of_study)):
    for j in range(len(gpas)):
        ax.text(j, i, f"{depression_scores[i,j]:.2f}", va='center', ha='center')

# Set axis labels
ax.set_xticks(np.arange(len(gpas)))
ax.set_yticks(np.arange(len(hours_of_study)))

# Set the axis labels
ax.set_xticklabels(gpas)
ax.set_yticklabels(hours_of_study)

# Rotate the tick labels for better readability
plt.xticks(rotation=45)

# Set axis labels
ax.set_xlabel('GPA')
ax.set_ylabel('Hours of Daily Work/Study')

plt.show()
```
